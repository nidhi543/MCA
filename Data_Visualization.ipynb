```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv("/tips (1).csv")
plt.scatter(data['day'],data['tip'])
plt.title("Scatter plot")
plt.xlabel('Day')
plt.ylabel('Tip')
plt.colorbar()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv("/tips (1).csv")
plt.plot(data['tip'])
plt.plot(data['size'])
plt.title("Scatter plot")
plt.xlabel('Day')
plt.ylabel('Tip')
plt.show()
```
