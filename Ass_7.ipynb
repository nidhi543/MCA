```python
import matplotlib.pyplot as plt
# x=[2,4,6,8]
# y=[3,7,5,10]
x = [4,5,10,4,3,11,14,6,10,12]
#  1 * 10

y = [21,19,24,17,16,25,24,22,21,21]
sum_x=sum(x)
sum_y=sum(y)
n=len(x)
xy=0
for i in range(len(x)):
  xy+= x[i]*y[i]
x_2=0
for i in range(len(x)):
  x_2+=x[i]**2
b=(n*xy - sum_x*sum_y)/(n*x_2 -sum_x**2)
a=((sum_y*x_2) - (sum_x*xy))/(n*x_2-sum_x**2)
print(a)
print(b)

def SLR(x):return (b*x+a)
x_new=[5,37,8,12,3,11,10]
New_X =[6,7,8,3,1,9,16,5,9,9,6,8,14,8,5,14,18,10,14,15]
New_Y = [22,18,25,15,14,26,22 ,20 ,24 ,19 ,27 ,25 ,27 ,22 ,21 ,28 ,29 ,25 ,25 ,27]
y_pred=[]
for i in New_X:
  ans=SLR(i)
  y_pred.append(ans)
plt.scatter(New_X,New_Y)
plt.plot(New_X,y_pred)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear regression')
```

```python
# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
np.random.seed(0)
X = 2 * np.random.rand(100, 1)   # Independent variable
y = 3 + 4 * X + np.random.randn(100, 1)   # Dependent variable

plt.scatter(X, y)
plt.xlabel('X')
plt.ylabel('y')
plt.title('Sample Data for Linear Regression')
plt.show()

model = LinearRegression()

model.fit(X, y)

slope = model.coef_[0][0]
intercept = model.intercept_[0]

print("Slope:", slope)
print("Intercept:", intercept)

# Making predictions
X_new = np.array([[0], [2]])
y_pred = model.predict(X_new)

plt.scatter(X, y)
plt.plot(X_new, y_pred, 'r-')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression')
plt.show()
```
