```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import pandas as pd
data1 = pd.read_csv('/content/data.csv')
x= data1['Duration']
y = data1['Calories']
data = list(zip(x,y))
print(data)
kmeans = KMeans(n_clusters=3)
kmeans.fit(data)
plt.scatter(x, y, c=kmeans.labels_)
plt.show()
```
